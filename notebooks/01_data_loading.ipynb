# 01. Data Loading and Consolidation
Load all TIKI fashion product CSV files and combine into unified dataset.

```python
import pandas as pd
import numpy as np
from pathlib import Path

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
```

## Load CSV Files

```python
dataset_path = Path('../dataset')

files = {
    'backpacks_suitcases': 'vietnamese_tiki_products_backpacks_suitcases.csv',
    'fashion_accessories': 'vietnamese_tiki_products_fashion_accessories.csv',
    'men_bags': 'vietnamese_tiki_products_men_bags.csv',
    'men_shoes': 'vietnamese_tiki_products_men_shoes.csv',
    'women_bags': 'vietnamese_tiki_products_women_bags.csv',
    'women_shoes': 'vietnamese_tiki_products_women_shoes.csv'
}

dataframes = {}
for category, filename in files.items():
    df = pd.read_csv(dataset_path / filename)
    df['category'] = category
    dataframes[category] = df
    print(f"{category}: {len(df)} rows, {len(df.columns)} columns")
```

## Combine All Datasets

```python
df_combined = pd.concat(dataframes.values(), ignore_index=True)
print(f"\nCombined dataset: {len(df_combined)} rows, {len(df_combined.columns)} columns")
print(f"\nCategory distribution:")
print(df_combined['category'].value_counts())
```

## Data Overview

```python
print("Column names:")
print(df_combined.columns.tolist())
print("\nData types:")
print(df_combined.dtypes)
print("\nFirst few rows:")
df_combined.head()
```

## Check for Duplicates

```python
duplicates = df_combined.duplicated(subset=['id']).sum()
print(f"Duplicate product IDs: {duplicates}")

if duplicates > 0:
    df_combined = df_combined.drop_duplicates(subset=['id'], keep='first')
    print(f"After removing duplicates: {len(df_combined)} rows")
```

## Save Combined Dataset

```python
output_path = Path('../data/processed')
output_path.mkdir(parents=True, exist_ok=True)

df_combined.to_csv(output_path / 'combined_tiki_data.csv', index=False)
print(f"Saved to {output_path / 'combined_tiki_data.csv'}")
```
